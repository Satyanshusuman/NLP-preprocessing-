```python
import numpy as np
import pandas as pd
import nltk
import string
```

```python
data =pd.read_csv("/content/sample_data/news.csv")
```

```python
df=pd.DataFrame(data["content"]).rename(columns={"content":"original content"})
df.dropna(inplace=True)
df
```

# **Step1: convert to lower case**

```python
s1=df["original content"].str.lower()
s1
```

# **step2: Tokenization**

```python
from nltk.tokenize import sent_tokenize
s2=s1.apply(lambda row:nltk.sent_tokenize(str(row)))
s2[809]
```

# **step3: Remove Punctuation**

```python
string.punctuation
```

```python
import re 
regex=re.compile("[%s]"% re.escape(string.punctuation))
def remove_punc(row): 
    new_list=[]
    for sent_token in row:
        new_token=regex.sub(u"",sent_token)
        if not new_token==u"":
            new_list.append(new_token)  
    return new_list

s3=s2.apply(remove_punc)
s3
```

# **Removing Stopwords**

```python
from nltk.corpus import stopwords
lsw=stopwords.words("english")
```

```python
from nltk.corpus import stopwords
def remove_stopwords(row): 
    new_list=[]
    for sent_token in row:
        for i in sent_token.split():
            if not i in lsw:
                new_list.append(i)     
    return " ".join(new_list)
        
s3.apply(remove_stopwords)
```
